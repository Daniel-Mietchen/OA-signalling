```python
import pywikibot
import mwparserfromhell
import json
import urllib2
import re
```

```python
enwp = pywikibot.Site('en', 'wikipedia')
doi_template = pywikibot.Page(enwp, 'Template:Cite doi')
doi_pages = doi_template.getReferences(follow_redirects=True, onlyTemplateInclusion=True, namespaces=0)
doi_page_objs = list(doi_pages)
doi_page_titles = map(lambda p: p.title(), doi_page_objs)
```

```python
def get_wvs(page_title):
    try:
        wvspath = 'https://tools.wmflabs.org/wikiviewstats/?locale=en&lang=en&project=&datefrom=2014-03-01&dateto=2014-04-01&page='
        path = wvspath + urllib2.quote(page_title)
        data = urllib2.urlopen(path)
        htmltext = data.read()
        match = re.search('([\d,]*) hits', htmltext).group(1)
        views = int(match.replace(',',''))
        return views
    except:
        return -1
```

```python
def get_wvs_crash(page_title):
    try:
        wvspath = 'https://tools.wmflabs.org/wikiviewstats/?locale=en&lang=en&project=&datefrom=2014-03-01&dateto=2014-04-01&page='
        path = wvspath + page_title.encode('utf-8')
        safepath = urllib2.quote(path, safe=";/?:@&=+$,")
        data = urllib2.urlopen(safepath)
        htmltext = data.read()
        dumpfile = open('dumpfile.html','w')
        dumpfile.write(htmltext)
        match = re.search('([\d,]*) hits', htmltext).group(1)
        views = int(match.replace(',',''))
        return views
    except:
        return -1
```

```python
page_views = dict()
for doi_page_title in doi_page_titles:
    try:
        print(page_views[doi_page_title])
    except:
        print('doing', doi_page_titles.index(doi_page_title), 'of', len(doi_page_titles))
        page_views[doi_page_title] = get_wvs(doi_page_title) 
        print(doi_page_title, page_views[doi_page_title])

f = open('page_views.json', 'w')
json.dump(page_views,f)
!du -h page_views.json
```

```python
def get_total_views(page_title, attempt=0):
    try:
        fullurl = u'http://stats.grok.se/json/en/200712/'+page_title.encode('utf-8')
        data = urllib2.urlopen(fullurl)
        jsondata = json.load(data)
        daily_views = jsondata['daily_views']
        total_views = reduce(lambda a,b: a+b, daily_views.itervalues())
    except urllib2.HTTPError:
        if attempt < 5:
            print('retrying', page_title)
            time.sleep(3)
            return get_total_views(page_title, attempt=(attempt+1))
        else:
            print('exception on', page_title)
            return 0
    return total_views
```
